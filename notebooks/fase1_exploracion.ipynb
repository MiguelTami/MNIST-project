The goal of this project is classify a set of images (MNIST) using multiple approaches, starting with hand-crafted features and progressing to neural networks.
 

```python
#En este punto asignamos nuestros dataframe a la variable data #

import pandas as pd

data = pd.read_csv("../data/raw/mnist_train.csv", header=None)
```

```python
#usamos .iloc y le asignamos a nuestra varaible X los valores que definen nuestro número Y y covertimos la data en array NumPy. 
#En este caso X corresponde a los 784 pixeles que conforman nuestro número Y

X = data.iloc[:, 1:].values
Y = data.iloc[:, 0].values
```

```python
#Importamos matplotlib para visualizar nuestro número 

import matplotlib.pyplot as plt
import matplotlib as mpl

n = 153

plt.imshow(X[n].reshape(28, 28), cmap=mpl.cm.binary)
plt.title(Y[n])
plt.show()
```

```python
pixel_values = data.iloc[:, 1:]  # sin la columna label
print("Valor de los pixeles:",pixel_values)
print("Media de pixeles:", pixel_values.values.mean())
print("Desviación estándar de pixeles:", pixel_values.values.std())
print("Valor máximo:", pixel_values.values.max())
print("Valor mínimo:", pixel_values.values.min())
```
